```python
# importing necessary libraries for data cleaning
import numpy as np
import pandas as pd
```

```python
# importing the data
df = pd.read_csv("/content/Airport_Project_Data.csv")

# keeping only useful columns
keep_cols = [
    "MAIL", "FREIGHT", "PASSENGERS",
    "DISTANCE_GROUP", "DISTANCE",
    "DEST_AIRPORT_ID", "DEST_CITY_NAME", "DEST",
    "ORIGIN_AIRPORT_ID", "ORIGIN_CITY_NAME", "ORIGIN",
    "UNIQUE_CARRIER_NAME", "UNIQUE_CARRIER",
    "MONTH", "QUARTER", "YEAR"
]

# copying useful columns into a new dataframe
df_cleaned = df[keep_cols].copy()

# filling missing values of freight column with 0
df_cleaned["FREIGHT"] = df_cleaned["FREIGHT"].fillna(0)

# preview
df_cleaned.head(10)
```

```python
# saving to CSV
df_cleaned.to_csv("Cleaned_Airport_Data.csv", index=False)

# downloading the cleaned dataset
from google.colab import files
files.download("Cleaned_Airport_Data.csv")
```
